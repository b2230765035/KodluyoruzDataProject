###  Gerekli Kütüphanelerin Yüklenmesi

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

### Veri Setlerinin Yüklenmesi ve İlk Temizlik
Veriler iki kaynaktan alınmıştır:
- **Kanser ölümleri verisi** (`Cancer Deaths by Country and Type Dataset.csv`)
- **Nüfus verisi** (`population.csv`)

Kodda ilk olarak veriler okunmuş ve eksik/boş satırlar temizlenmiştir.

```python
data = pd.read_csv('cancerdeaths.csv')
population = pd.read_csv("population.csv")

# "Code" sütununda eksik olan satırlar veri dışı bırakılıyor
data.dropna(subset=["Code"], inplace=True)
data = data.reset_index(drop=True)

# 1990 ve sonrasındaki yıllar filtreleniyor
population = population[population["Year"] >= 1990]
population = population.rename(columns={"Entity": "Country", "Population (historical)": "Population"})
population["Year"] = population["Year"].astype(str)
```

### Kanser Oranı Hesaplama Fonksiyonu

Her ülke ve yıl için toplam kanser vakası hesaplanarak, nüfusa oranla **100.000 kişide görülen kanser oranı** (`Cancer_per_100k`) elde edilmiştir. Bu hesaplamalar daha sağlıklı karşılaştırmalar yapmamızı sağlar..


```python
def hesapla_kanser_orani(df,nufus):
    grouped = df.groupby(["Country", "Year"]).sum(numeric_only=True)#yıl+ülkeye göre grupladık ve kanser vakalrını topladık
    grouped["Total_Cancer"] = grouped.sum(axis=1)
    grouped = grouped.reset_index()
    grouped["Year"] = grouped["Year"].astype(str)
    merged = pd.merge(grouped, nufus, on=["Country", "Year"], how="left")#Toplam kanser verisi ile nüfus verisi "Country" ve "Year" üzerinden birleştirildi
    merged = merged.dropna(subset=["Population"])
    merged["Cancer_per_100k"] = (merged["Total_Cancer"] / merged["Population"]) * 100000
    return merged

merged = hesapla_kanser_orani(data,population)
```

### Yıllara Göre Toplam Kanser Vakası (Dünya)

Bu grafikte her yıl dünya genelinde kaydedilen toplam kanser vakalarının değişimi gösterilmiştir. Zaman içinde artış ya da azalış eğilimleri analiz edilebilir.

```python
#Her yıl tüm dünyadaki toplam kanser oranları ne kadar? Artıyor mu, azalıyor mu?
#Yıl ekseninde değişimi daha iyi göstermesi için line plot kullanıldı
plt.figure(figsize=(10, 5))
merged.groupby("Year")["Total_Cancer"].sum().plot()
plt.title("Yıllara Göre Toplam Kanser Vakası (Dünya)")
plt.ylabel("Toplam Kanser Vakası")
plt.xlabel("Yıl")
plt.grid()
plt.tight_layout()
plt.show()
```

### En Fazla Kanser Vakası Görülen Ülkeler (Top 10)

Tüm yıllar boyunca toplam kanser vakası en yüksek olan ilk 10 ülke yatay çubuk grafik ile görselleştirilmiştir. Bu ülkelerdeki sağlık sistemleri, yaşam tarzları veya çevresel faktörler araştırılabilir.r.

```python
#Hangi ülkede toplamda daha fazla kanser var?
top10 = merged.groupby("Country")["Total_Cancer"].sum().sort_values(ascending=False).head(10)
plt.figure(figsize=(10, 6))
top10.plot(kind="barh")
plt.title("Toplam Kanser Vakası (En Yüksek 10 Ülke)")
plt.xlabel("Toplam Kanser Vakası")
plt.grid(axis='x')
plt.tight_layout()
plt.show()
```

###  Yıllara Göre En Yüksek Kanser Oranına Sahip İlk 3 Ülke

Her yıl kişi başına düşen kanser oranı en yüksek olan 3 ülke seçilmiş ve bu ülkelerin yıllara göre değişimi görselleştirilmiştir. Amaç; hangi ülkelerde kanser oranlarının daha yüksek seyrettiğini analiz etmektir.
.


```python
# Her yıl için en yüksek kanser oranına sahip ilk 3 ülkeyi bulduk
top3_filtered = (
    merged.sort_values(["Year", "Cancer_per_100k"], ascending=[True, False])
    .groupby("Year")
    .head(3)
)

# Pivot tablo
pivot_df = top3_filtered.pivot(index="Year", columns="Country", values="Cancer_per_100k")

pivot_df.plot(kind="line", marker="o", figsize=(12, 6),
              title="Yıllara Göre En Fazla Kanser Oranına Sahip 3 Gerçek Ülke")
plt.ylabel("Toplam Kanser Oranı")
plt.xlabel("Yıl")
plt.grid(True)
plt.legend(title="Ülke")
plt.tight_layout()
plt.show()
```

### Ülkelere Göre Kanser Oranındaki Sapmalar

Her ülkenin yıllık toplam kanser vakası, o yılın genel ortalamasıyla karşılaştırılmıştır. Bu sayede aşağıdaki üç kategoriye göre analiz yapılmıştır:

- 🔴 Ortalama Üzerinde: Sürekli yüksek vaka sayısı olan ülkeler
- 🟢 Ortalama Altında: Sürekli düşük vaka sayısı olan ülkeler
- 🟡 Dalgalı: Ortalama etrafında dalgalanan ülkeler

Bu analiz, zaman içinde sağlıkta iyileşme veya kötüleşme sinyallerini anlamaya yardımcı olur.


```python
#Her ülkenin ortalama yıllık kanser oranı ne? Genel ortalamadan ne kadar sapıyor?
#Yani sağlıkta iyileşme veya kötüleşme analiz edilebilir

population["Country"] = population["Country"].str.strip().str.lower()

data["Country"] = data["Country"].str.strip().str.lower()
data["Year"] = data["Year"].astype(str)
pop_lookup = population.set_index(["Country", "Year"])["Population"].to_dict()
data["Population"] = data.set_index(["Country", "Year"]).index.map(pop_lookup)
cancer_columns = data.columns[3:-1]
data["Total_Cancer"] = data[cancer_columns].sum(axis=1)
data["Cancer_per_100k"] = (data["Total_Cancer"] / data["Population"]) * 100000

# Yıl ve ülkeye göre toplam
grouped = data.groupby(["Year", "Country"]).sum(numeric_only=True)
grouped["Total"] = grouped["Total_Cancer"]

# Yıllık ortalama ve sapma
yearly_avg = grouped["Total"].groupby("Year").mean()
grouped = grouped.reset_index()
grouped["Sapma"] = grouped.apply(lambda row: row["Total"] - yearly_avg[row["Year"]], axis=1)

# Her ülkenin tüm yıllar için ortalama sapması hesaplandı
country_sapma_avg = grouped.groupby("Country")["Sapma"].mean()

high = country_sapma_avg[country_sapma_avg > 500].index.tolist()
low = country_sapma_avg[country_sapma_avg < -500].index.tolist()
mid = country_sapma_avg[(country_sapma_avg >= -500) & (country_sapma_avg <= 500)].index.tolist()

def draw_sapma_plot_for(countries, title):
    plt.figure(figsize=(12, 6))
    for country in countries:
        df = grouped[grouped["Country"] == country]
        plt.plot(df["Year"], df["Sapma"], label=country.title(), marker='o')
    plt.axhline(0, color="gray", linestyle="--")
    plt.title(title)
    plt.xlabel("Yıl")
    plt.ylabel("Toplam Kanser Sapması")
    plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', fontsize=8)
    plt.tight_layout()
    plt.grid(True)
    plt.show()

# Grafikleri çiz
draw_sapma_plot_for(high[:6], "🔴 Sürekli Ortalamanın Üzerinde Olan Ülkeler")
draw_sapma_plot_for(low[:6], "🟢 Sürekli Ortalamanın Altında Olan Ülkeler")
draw_sapma_plot_for(mid[:6], "🟡 Ortalama Etrafında Dalgalanan Ülkeler")
```

### Yıllara Göre En Yüksek Kanser Oranına Sahip Ülkeler ve En Yaygın Kanser Türü

Bu analizde, her yıl kişi başına en yüksek kanser oranına sahip ülke belirlenmiş ve bu ülkelerdeki en yaygın kanser türü tespit edilmiştir. Amaç; en çok etkilenen ülkeyi ve hangi kanser türünün öne çıktığını yıllık bazda ortaya koymaktır.

- Her bir yıl için **en yüksek 'Cancer_per_100k' oranına sahip ülke** seçilmiştir.
- Seçilen ülke için **o yıl en yaygın olan kanser türü** tespit edilmiştir.
- Sonuçlar **yıllık çubuk grafik** ile görselleştirilmiş, her çubuğa ülke adı eklenmiştir.
- Renkler, baskın kanser türlerini temsil eder.


```python
data.columns = data.columns.str.strip()
cancer_columns = data.columns[3:-3]
cancer_columns = [col.strip() for col in cancer_columns]  # Boşlukları temizle

data["Total_Cancer"] = data[cancer_columns].sum(axis=1)
data["Cancer_per_100k"] = (data["Total_Cancer"] / data["Population"]) * 100000

# Her yıl kişi başına en yüksek kanser oranına sahip ülkeyi bul
top_country_per_year = (
    data.loc[data.groupby("Year")["Cancer_per_100k"].idxmax()]
    .reset_index(drop=True)
)

# Her yıl için o ülkenin en yaygın kanser türünü bul
most_common_types = []

for _, row in top_country_per_year.iterrows():
    year = row["Year"]
    country = row["Country"].strip().lower()

    row_data = data[(data["Year"] == year) & (data["Country"] == country)]

    if not row_data.empty:
        max_cancer_type = row_data[cancer_columns].iloc[0].idxmax()
        most_common_types.append(max_cancer_type)
    else:
        most_common_types.append("Bilinmiyor")

top_country_per_year["Most_Common_Cancer"] = most_common_types
top_country_per_year["Country"] = top_country_per_year["Country"].str.title()

plt.figure(figsize=(14, 6))
sns.barplot(data=top_country_per_year, x="Year", y="Cancer_per_100k", hue="Most_Common_Cancer", dodge=False)

# Her çubuğun üzerine ülke adını yaz
for i, row in top_country_per_year.iterrows():
    label = f"{row['Country']}"
    plt.text(x=i, y=row["Cancer_per_100k"] + 1, s=label, ha='center', fontsize=7, rotation=90)

plt.title("Yıllara Göre Kişi Başına En Yüksek Kanser Oranına Sahip Ülke ve Baskın Kanser Türü")
plt.ylabel("Kanser Oranı (100k Kişi Başına)")
plt.xlabel("Yıl")
plt.xticks(rotation=45)
plt.legend(title="Baskın Kanser Türü", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

## Kıta Bilgisinin Eklenmesi  
Ülkelere ait kıta bilgileri eklendi ve veriler temizlendi. Böylece coğrafi temelli analizler ve makine öğrenimi modelleri için daha anlamlı girdiler elde edildi.

```python
# Kıta bilgisi ile veri setini birleştirelim
continent_df = pd.read_csv("Countries by continents.csv")
continent_df["Country"] = continent_df["Country"].str.strip().str.lower()
merged["Country"] = merged["Country"].str.strip().str.lower()

# Kıta bazında risk sınıflandırması için birleşim
merged_with_continent = pd.merge(merged, continent_df, on="Country", how="left")
cleaned = merged_with_continent.dropna(subset=["Continent"]).copy()

```

###  Kanser Risk Seviyelerinin Sınıflandırılması  
Toplam kanser vakasına göre ülkeler "Low", "Medium" ve "High" olmak üzere üç risk seviyesine ayrıldı. Bu sınıflandırma makine öğrenimi modellerinde hedef değişken olarak kullanıldı.

```python
# Risk seviyesini tanımlayan fonksiyon
def yeni_risk_sinifi(total):
    if total < 1500:
        return "Low"
    elif total < 4000:
        return "Medium"
    else:
        return "High"

cleaned["Risk_Level"] = cleaned["Total_Cancer"].apply(yeni_risk_sinifi)
```

###  Eğitim ve Test Verisinin Zaman Bazlı Ayrımı (Train-Test Split)


```python
# Train-test ayırımı (2014 yılına kadar eğitim, sonrası test)
train_mask = cleaned["Year"].astype(int) <= 2014
test_mask = cleaned["Year"].astype(int) > 2014
```

### Coğrafi Temelli Özelliklerle Model Girdilerinin Hazırlanması ve Ölçeklendirme


```python
# Bağımsız değişkenler: Nüfus + kıta bilgisi (One-hot encoding)
X_geo = pd.concat([
    cleaned[["Population"]], 
    pd.get_dummies(cleaned["Continent"], prefix="Continent")
], axis=1)
y_geo = cleaned["Risk_Level"]

X_train_g, X_test_g = X_geo[train_mask], X_geo[test_mask]
y_train_g, y_test_g = y_geo[train_mask], y_geo[test_mask]

#Ölçeklendirme
scaler_g = StandardScaler()
X_train_g_scaled = scaler_g.fit_transform(X_train_g)
X_test_g_scaled = scaler_g.transform(X_test_g)
```

### Random Forest Modeli ile Kanser Risk Seviyesi Sınıflandırması ve Performans Değerlendirmesi


```python
rf_geo = RandomForestClassifier(random_state=42)
rf_geo.fit(X_train_g_scaled, y_train_g)
y_pred_rf = rf_geo.predict(X_test_g_scaled)

print("🔍 Random Forest")
print("Accuracy:", accuracy_score(y_test_g, y_pred_rf))
print("Confusion Matrix:\n", confusion_matrix(y_test_g, y_pred_rf))
print("Classification Report:\n", classification_report(y_test_g, y_pred_rf))
rf_geo = RandomForestClassifier(random_state=42)
rf_geo.fit(X_train_g_scaled, y_train_g)
y_pred_rf = rf_geo.predict(X_test_g_scaled)

print("🔍 Random Forest")
print("Accuracy:", accuracy_score(y_test_g, y_pred_rf))
print("Confusion Matrix:\n", confusion_matrix(y_test_g, y_pred_rf))
print("Classification Report:\n", classification_report(y_test_g, y_pred_rf))
```

#### Random Forest Modeli Sonuçları

Model %90.5 doğrulukla iyi performans gösteriyor. Yüksek ve düşük risk sınıflarını başarıyla ayırırken, orta risk sınıfında daha zayıf kalıyor. Orta risk için model geliştirmeye ihtiyaç var.


### K-Nearest Neighbors (KNN) Modeli ile Risk Seviyesi Sınıflandırması


```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train_g_scaled, y_train_g)
y_pred_knn = knn.predict(X_test_g_scaled)

print("🔍 K-Nearest Neighbors")
print("Accuracy:", accuracy_score(y_test_g, y_pred_knn))
print("Confusion Matrix:\n", confusion_matrix(y_test_g, y_pred_knn))
print("Classification Report:\n", classification_report(y_test_g, y_pred_knn))
```

#### K-Nearest Neighbors Modeli Sonuçları ve Kısa Değerlendirme

Model genel olarak %91 doğruluk oranı ile başarılı sonuç verdi.  
Yüksek ve düşük risk sınıflarını iyi tahmin ederken, orta risk sınıfında performans biraz daha düşük kaldı.  
Precision ve recall dengesi makul seviyede, orta riskte iyileştirme gerekebilir.


### Support Vector Machine (SVC) Modeli Eğitimi 


```python
from sklearn.svm import SVC

svc = SVC()
svc.fit(X_train_g_scaled, y_train_g)
y_pred_svc = svc.predict(X_test_g_scaled)

print("🔍 Support Vector Machine (SVC)")
print("Accuracy:", accuracy_score(y_test_g, y_pred_svc))
print("Confusion Matrix:\n", confusion_matrix(y_test_g, y_pred_svc))
print("Classification Report:\n", classification_report(y_test_g, y_pred_svc))
```

#### Support Vector Machine (SVC) Modeli Sonuçları

Model genel doğruluk oranı %79 olup, yüksek ve düşük risk sınıflarında makul performans gösterirken, orta risk sınıfını hiç doğru tahmin edememiştir. Bu durum, modelin sınıflar arasında dengesiz kaldığını ve orta risk sınıfını ayırt etmekte zorlandığını göstermektedir.


### Logistic Regression Modeli

#### Logistic Regression Sonuç Yorumu

### Decision Tree Classifier

```python
from sklearn.tree import DecisionTreeClassifier

dt = DecisionTreeClassifier(random_state=42)
dt.fit(X_train_g_scaled, y_train_g)
y_pred_dt = dt.predict(X_test_g_scaled)

print("🔍 Decision Tree Classifier")
print("Accuracy:", accuracy_score(y_test_g, y_pred_dt))
print("Confusion Matrix:\n", confusion_matrix(y_test_g, y_pred_dt))
print("Classification Report:\n", classification_report(y_test_g, y_pred_dt))
```
