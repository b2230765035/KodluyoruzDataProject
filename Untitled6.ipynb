###  Gerekli KÃ¼tÃ¼phanelerin YÃ¼klenmesi

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

### Veri Setlerinin YÃ¼klenmesi ve Ä°lk Temizlik
Veriler iki kaynaktan alÄ±nmÄ±ÅŸtÄ±r:
- **Kanser Ã¶lÃ¼mleri verisi** (`Cancer Deaths by Country and Type Dataset.csv`)
- **NÃ¼fus verisi** (`population.csv`)

Kodda ilk olarak veriler okunmuÅŸ ve eksik/boÅŸ satÄ±rlar temizlenmiÅŸtir.

```python
data = pd.read_csv('cancerdeaths.csv')
population = pd.read_csv("population.csv")

# "Code" sÃ¼tununda eksik olan satÄ±rlar veri dÄ±ÅŸÄ± bÄ±rakÄ±lÄ±yor
data.dropna(subset=["Code"], inplace=True)
data = data.reset_index(drop=True)

# 1990 ve sonrasÄ±ndaki yÄ±llar filtreleniyor
population = population[population["Year"] >= 1990]
population = population.rename(columns={"Entity": "Country", "Population (historical)": "Population"})
population["Year"] = population["Year"].astype(str)
```

### Kanser OranÄ± Hesaplama Fonksiyonu

Her Ã¼lke ve yÄ±l iÃ§in toplam kanser vakasÄ± hesaplanarak, nÃ¼fusa oranla **100.000 kiÅŸide gÃ¶rÃ¼len kanser oranÄ±** (`Cancer_per_100k`) elde edilmiÅŸtir. Bu hesaplamalar daha saÄŸlÄ±klÄ± karÅŸÄ±laÅŸtÄ±rmalar yapmamÄ±zÄ± saÄŸlar..


```python
def hesapla_kanser_orani(df,nufus):
    grouped = df.groupby(["Country", "Year"]).sum(numeric_only=True)#yÄ±l+Ã¼lkeye gÃ¶re grupladÄ±k ve kanser vakalrÄ±nÄ± topladÄ±k
    grouped["Total_Cancer"] = grouped.sum(axis=1)
    grouped = grouped.reset_index()
    grouped["Year"] = grouped["Year"].astype(str)
    merged = pd.merge(grouped, nufus, on=["Country", "Year"], how="left")#Toplam kanser verisi ile nÃ¼fus verisi "Country" ve "Year" Ã¼zerinden birleÅŸtirildi
    merged = merged.dropna(subset=["Population"])
    merged["Cancer_per_100k"] = (merged["Total_Cancer"] / merged["Population"]) * 100000
    return merged

merged = hesapla_kanser_orani(data,population)
```

### YÄ±llara GÃ¶re Toplam Kanser VakasÄ± (DÃ¼nya)

Bu grafikte her yÄ±l dÃ¼nya genelinde kaydedilen toplam kanser vakalarÄ±nÄ±n deÄŸiÅŸimi gÃ¶sterilmiÅŸtir. Zaman iÃ§inde artÄ±ÅŸ ya da azalÄ±ÅŸ eÄŸilimleri analiz edilebilir.

```python
#Her yÄ±l tÃ¼m dÃ¼nyadaki toplam kanser oranlarÄ± ne kadar? ArtÄ±yor mu, azalÄ±yor mu?
#YÄ±l ekseninde deÄŸiÅŸimi daha iyi gÃ¶stermesi iÃ§in line plot kullanÄ±ldÄ±
plt.figure(figsize=(10, 5))
merged.groupby("Year")["Total_Cancer"].sum().plot()
plt.title("YÄ±llara GÃ¶re Toplam Kanser VakasÄ± (DÃ¼nya)")
plt.ylabel("Toplam Kanser VakasÄ±")
plt.xlabel("YÄ±l")
plt.grid()
plt.tight_layout()
plt.show()
```

### En Fazla Kanser VakasÄ± GÃ¶rÃ¼len Ãœlkeler (Top 10)

TÃ¼m yÄ±llar boyunca toplam kanser vakasÄ± en yÃ¼ksek olan ilk 10 Ã¼lke yatay Ã§ubuk grafik ile gÃ¶rselleÅŸtirilmiÅŸtir. Bu Ã¼lkelerdeki saÄŸlÄ±k sistemleri, yaÅŸam tarzlarÄ± veya Ã§evresel faktÃ¶rler araÅŸtÄ±rÄ±labilir.r.

###  YÄ±llara GÃ¶re En YÃ¼ksek Kanser OranÄ±na Sahip Ä°lk 3 Ãœlke

Her yÄ±l kiÅŸi baÅŸÄ±na dÃ¼ÅŸen kanser oranÄ± en yÃ¼ksek olan 3 Ã¼lke seÃ§ilmiÅŸ ve bu Ã¼lkelerin yÄ±llara gÃ¶re deÄŸiÅŸimi gÃ¶rselleÅŸtirilmiÅŸtir. AmaÃ§; hangi Ã¼lkelerde kanser oranlarÄ±nÄ±n daha yÃ¼ksek seyrettiÄŸini analiz etmektir.
.


```python
# Her yÄ±l iÃ§in en yÃ¼ksek kanser oranÄ±na sahip ilk 3 Ã¼lkeyi bulduk
top3_filtered = (
    merged.sort_values(["Year", "Cancer_per_100k"], ascending=[True, False])
    .groupby("Year")
    .head(3)
)

# Pivot tablo
pivot_df = top3_filtered.pivot(index="Year", columns="Country", values="Cancer_per_100k")

pivot_df.plot(kind="line", marker="o", figsize=(12, 6),
              title="YÄ±llara GÃ¶re En Fazla Kanser OranÄ±na Sahip 3 GerÃ§ek Ãœlke")
plt.ylabel("Toplam Kanser OranÄ±")
plt.xlabel("YÄ±l")
plt.grid(True)
plt.legend(title="Ãœlke")
plt.tight_layout()
plt.show()
```

### Ãœlkelere GÃ¶re Kanser OranÄ±ndaki Sapmalar

Her Ã¼lkenin yÄ±llÄ±k toplam kanser vakasÄ±, o yÄ±lÄ±n genel ortalamasÄ±yla karÅŸÄ±laÅŸtÄ±rÄ±lmÄ±ÅŸtÄ±r. Bu sayede aÅŸaÄŸÄ±daki Ã¼Ã§ kategoriye gÃ¶re analiz yapÄ±lmÄ±ÅŸtÄ±r:

- ðŸ”´ Ortalama Ãœzerinde: SÃ¼rekli yÃ¼ksek vaka sayÄ±sÄ± olan Ã¼lkeler
- ðŸŸ¢ Ortalama AltÄ±nda: SÃ¼rekli dÃ¼ÅŸÃ¼k vaka sayÄ±sÄ± olan Ã¼lkeler
- ðŸŸ¡ DalgalÄ±: Ortalama etrafÄ±nda dalgalanan Ã¼lkeler

Bu analiz, zaman iÃ§inde saÄŸlÄ±kta iyileÅŸme veya kÃ¶tÃ¼leÅŸme sinyallerini anlamaya yardÄ±mcÄ± olur.


```python
#Her Ã¼lkenin ortalama yÄ±llÄ±k kanser oranÄ± ne? Genel ortalamadan ne kadar sapÄ±yor?
#Yani saÄŸlÄ±kta iyileÅŸme veya kÃ¶tÃ¼leÅŸme analiz edilebilir

population["Country"] = population["Country"].str.strip().str.lower()

data["Country"] = data["Country"].str.strip().str.lower()
data["Year"] = data["Year"].astype(str)
pop_lookup = population.set_index(["Country", "Year"])["Population"].to_dict()
data["Population"] = data.set_index(["Country", "Year"]).index.map(pop_lookup)
cancer_columns = data.columns[3:-1]
data["Total_Cancer"] = data[cancer_columns].sum(axis=1)
data["Cancer_per_100k"] = (data["Total_Cancer"] / data["Population"]) * 100000

# YÄ±l ve Ã¼lkeye gÃ¶re toplam
grouped = data.groupby(["Year", "Country"]).sum(numeric_only=True)
grouped["Total"] = grouped["Total_Cancer"]

# YÄ±llÄ±k ortalama ve sapma
yearly_avg = grouped["Total"].groupby("Year").mean()
grouped = grouped.reset_index()
grouped["Sapma"] = grouped.apply(lambda row: row["Total"] - yearly_avg[row["Year"]], axis=1)

# Her Ã¼lkenin tÃ¼m yÄ±llar iÃ§in ortalama sapmasÄ± hesaplandÄ±
country_sapma_avg = grouped.groupby("Country")["Sapma"].mean()

high = country_sapma_avg[country_sapma_avg > 500].index.tolist()
low = country_sapma_avg[country_sapma_avg < -500].index.tolist()
mid = country_sapma_avg[(country_sapma_avg >= -500) & (country_sapma_avg <= 500)].index.tolist()

def draw_sapma_plot_for(countries, title):
    plt.figure(figsize=(12, 6))
    for country in countries:
        df = grouped[grouped["Country"] == country]
        plt.plot(df["Year"], df["Sapma"], label=country.title(), marker='o')
    plt.axhline(0, color="gray", linestyle="--")
    plt.title(title)
    plt.xlabel("YÄ±l")
    plt.ylabel("Toplam Kanser SapmasÄ±")
    plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', fontsize=8)
    plt.tight_layout()
    plt.grid(True)
    plt.show()

# Grafikleri Ã§iz
draw_sapma_plot_for(high[:6], "ðŸ”´ SÃ¼rekli OrtalamanÄ±n Ãœzerinde Olan Ãœlkeler")
draw_sapma_plot_for(low[:6], "ðŸŸ¢ SÃ¼rekli OrtalamanÄ±n AltÄ±nda Olan Ãœlkeler")
draw_sapma_plot_for(mid[:6], "ðŸŸ¡ Ortalama EtrafÄ±nda Dalgalanan Ãœlkeler")
```

### YÄ±llara GÃ¶re En YÃ¼ksek Kanser OranÄ±na Sahip Ãœlkeler ve En YaygÄ±n Kanser TÃ¼rÃ¼

Bu analizde, her yÄ±l kiÅŸi baÅŸÄ±na en yÃ¼ksek kanser oranÄ±na sahip Ã¼lke belirlenmiÅŸ ve bu Ã¼lkelerdeki en yaygÄ±n kanser tÃ¼rÃ¼ tespit edilmiÅŸtir. AmaÃ§; en Ã§ok etkilenen Ã¼lkeyi ve hangi kanser tÃ¼rÃ¼nÃ¼n Ã¶ne Ã§Ä±ktÄ±ÄŸÄ±nÄ± yÄ±llÄ±k bazda ortaya koymaktÄ±r.

- Her bir yÄ±l iÃ§in **en yÃ¼ksek 'Cancer_per_100k' oranÄ±na sahip Ã¼lke** seÃ§ilmiÅŸtir.
- SeÃ§ilen Ã¼lke iÃ§in **o yÄ±l en yaygÄ±n olan kanser tÃ¼rÃ¼** tespit edilmiÅŸtir.
- SonuÃ§lar **yÄ±llÄ±k Ã§ubuk grafik** ile gÃ¶rselleÅŸtirilmiÅŸ, her Ã§ubuÄŸa Ã¼lke adÄ± eklenmiÅŸtir.
- Renkler, baskÄ±n kanser tÃ¼rlerini temsil eder.


```python
data.columns = data.columns.str.strip()
cancer_columns = data.columns[3:-3]
cancer_columns = [col.strip() for col in cancer_columns]  # BoÅŸluklarÄ± temizle

data["Total_Cancer"] = data[cancer_columns].sum(axis=1)
data["Cancer_per_100k"] = (data["Total_Cancer"] / data["Population"]) * 100000

# Her yÄ±l kiÅŸi baÅŸÄ±na en yÃ¼ksek kanser oranÄ±na sahip Ã¼lkeyi bul
top_country_per_year = (
    data.loc[data.groupby("Year")["Cancer_per_100k"].idxmax()]
    .reset_index(drop=True)
)

# Her yÄ±l iÃ§in o Ã¼lkenin en yaygÄ±n kanser tÃ¼rÃ¼nÃ¼ bul
most_common_types = []

for _, row in top_country_per_year.iterrows():
    year = row["Year"]
    country = row["Country"].strip().lower()

    row_data = data[(data["Year"] == year) & (data["Country"] == country)]

    if not row_data.empty:
        max_cancer_type = row_data[cancer_columns].iloc[0].idxmax()
        most_common_types.append(max_cancer_type)
    else:
        most_common_types.append("Bilinmiyor")

top_country_per_year["Most_Common_Cancer"] = most_common_types
top_country_per_year["Country"] = top_country_per_year["Country"].str.title()

plt.figure(figsize=(14, 6))
sns.barplot(data=top_country_per_year, x="Year", y="Cancer_per_100k", hue="Most_Common_Cancer", dodge=False)

# Her Ã§ubuÄŸun Ã¼zerine Ã¼lke adÄ±nÄ± yaz
for i, row in top_country_per_year.iterrows():
    label = f"{row['Country']}"
    plt.text(x=i, y=row["Cancer_per_100k"] + 1, s=label, ha='center', fontsize=7, rotation=90)

plt.title("YÄ±llara GÃ¶re KiÅŸi BaÅŸÄ±na En YÃ¼ksek Kanser OranÄ±na Sahip Ãœlke ve BaskÄ±n Kanser TÃ¼rÃ¼")
plt.ylabel("Kanser OranÄ± (100k KiÅŸi BaÅŸÄ±na)")
plt.xlabel("YÄ±l")
plt.xticks(rotation=45)
plt.legend(title="BaskÄ±n Kanser TÃ¼rÃ¼", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

## KÄ±ta Bilgisinin Eklenmesi  
Ãœlkelere ait kÄ±ta bilgileri eklendi ve veriler temizlendi. BÃ¶ylece coÄŸrafi temelli analizler ve makine Ã¶ÄŸrenimi modelleri iÃ§in daha anlamlÄ± girdiler elde edildi.

```python
# KÄ±ta bilgisi ile veri setini birleÅŸtirelim
continent_df = pd.read_csv("Countries by continents.csv")
continent_df["Country"] = continent_df["Country"].str.strip().str.lower()
merged["Country"] = merged["Country"].str.strip().str.lower()

# KÄ±ta bazÄ±nda risk sÄ±nÄ±flandÄ±rmasÄ± iÃ§in birleÅŸim
merged_with_continent = pd.merge(merged, continent_df, on="Country", how="left")
cleaned = merged_with_continent.dropna(subset=["Continent"]).copy()

```

###  Kanser Risk Seviyelerinin SÄ±nÄ±flandÄ±rÄ±lmasÄ±  
Toplam kanser vakasÄ±na gÃ¶re Ã¼lkeler "Low", "Medium" ve "High" olmak Ã¼zere Ã¼Ã§ risk seviyesine ayrÄ±ldÄ±. Bu sÄ±nÄ±flandÄ±rma makine Ã¶ÄŸrenimi modellerinde hedef deÄŸiÅŸken olarak kullanÄ±ldÄ±.

```python
# Risk seviyesini tanÄ±mlayan fonksiyon
def yeni_risk_sinifi(total):
    if total < 1500:
        return "Low"
    elif total < 4000:
        return "Medium"
    else:
        return "High"

cleaned["Risk_Level"] = cleaned["Total_Cancer"].apply(yeni_risk_sinifi)
```

###  EÄŸitim ve Test Verisinin Zaman BazlÄ± AyrÄ±mÄ± (Train-Test Split)


```python
# Train-test ayÄ±rÄ±mÄ± (2014 yÄ±lÄ±na kadar eÄŸitim, sonrasÄ± test)
train_mask = cleaned["Year"].astype(int) <= 2014
test_mask = cleaned["Year"].astype(int) > 2014
```

### CoÄŸrafi Temelli Ã–zelliklerle Model Girdilerinin HazÄ±rlanmasÄ± ve Ã–lÃ§eklendirme


```python
# BaÄŸÄ±msÄ±z deÄŸiÅŸkenler: NÃ¼fus + kÄ±ta bilgisi (One-hot encoding)
X_geo = pd.concat([
    cleaned[["Population"]], 
    pd.get_dummies(cleaned["Continent"], prefix="Continent")
], axis=1)
y_geo = cleaned["Risk_Level"]

X_train_g, X_test_g = X_geo[train_mask], X_geo[test_mask]
y_train_g, y_test_g = y_geo[train_mask], y_geo[test_mask]

#Ã–lÃ§eklendirme
scaler_g = StandardScaler()
X_train_g_scaled = scaler_g.fit_transform(X_train_g)
X_test_g_scaled = scaler_g.transform(X_test_g)
```

### Random Forest Modeli ile Kanser Risk Seviyesi SÄ±nÄ±flandÄ±rmasÄ± ve Performans DeÄŸerlendirmesi


```python
rf_geo = RandomForestClassifier(random_state=42)
rf_geo.fit(X_train_g_scaled, y_train_g)
y_pred_rf = rf_geo.predict(X_test_g_scaled)

print("ðŸ” Random Forest")
print("Accuracy:", accuracy_score(y_test_g, y_pred_rf))
print("Confusion Matrix:\n", confusion_matrix(y_test_g, y_pred_rf))
print("Classification Report:\n", classification_report(y_test_g, y_pred_rf))
rf_geo = RandomForestClassifier(random_state=42)
rf_geo.fit(X_train_g_scaled, y_train_g)
y_pred_rf = rf_geo.predict(X_test_g_scaled)

print("ðŸ” Random Forest")
print("Accuracy:", accuracy_score(y_test_g, y_pred_rf))
print("Confusion Matrix:\n", confusion_matrix(y_test_g, y_pred_rf))
print("Classification Report:\n", classification_report(y_test_g, y_pred_rf))
```

#### Random Forest Modeli SonuÃ§larÄ±

Model %90.5 doÄŸrulukla iyi performans gÃ¶steriyor. YÃ¼ksek ve dÃ¼ÅŸÃ¼k risk sÄ±nÄ±flarÄ±nÄ± baÅŸarÄ±yla ayÄ±rÄ±rken, orta risk sÄ±nÄ±fÄ±nda daha zayÄ±f kalÄ±yor. Orta risk iÃ§in model geliÅŸtirmeye ihtiyaÃ§ var.


### K-Nearest Neighbors (KNN) Modeli ile Risk Seviyesi SÄ±nÄ±flandÄ±rmasÄ±


```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train_g_scaled, y_train_g)
y_pred_knn = knn.predict(X_test_g_scaled)

print("ðŸ” K-Nearest Neighbors")
print("Accuracy:", accuracy_score(y_test_g, y_pred_knn))
print("Confusion Matrix:\n", confusion_matrix(y_test_g, y_pred_knn))
print("Classification Report:\n", classification_report(y_test_g, y_pred_knn))
```

#### K-Nearest Neighbors Modeli SonuÃ§larÄ± ve KÄ±sa DeÄŸerlendirme

Model genel olarak %91 doÄŸruluk oranÄ± ile baÅŸarÄ±lÄ± sonuÃ§ verdi.  
YÃ¼ksek ve dÃ¼ÅŸÃ¼k risk sÄ±nÄ±flarÄ±nÄ± iyi tahmin ederken, orta risk sÄ±nÄ±fÄ±nda performans biraz daha dÃ¼ÅŸÃ¼k kaldÄ±.  
Precision ve recall dengesi makul seviyede, orta riskte iyileÅŸtirme gerekebilir.


### Support Vector Machine (SVC) Modeli EÄŸitimi
